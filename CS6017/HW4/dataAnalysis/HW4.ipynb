```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
#task 1
dumb = pd.read_csv("dumbKnn.txt")
bucket = pd.read_csv("bucketKnn.txt")
quad = pd.read_csv("quadTreeKnn.txt")
kd = pd.read_csv("kdTreeKnn.txt")
```

```python
display(kd)
```

```python
start = 0
end = 9
print(dumb.iloc[end])
plt.scatter(x = dumb["dimension"].iloc[start:end], y = dumb["time"].iloc[start:end], color = "red")
plt.scatter(x = bucket["dimension"].iloc[start:end], y = bucket["time"].iloc[start:end], color = "blue")
plt.scatter(x = kd["dimension"].iloc[start:end], y = kd["time"].iloc[start:end], color = "black")
plt.legend(["dumb", "bucket", "kd"],)
plt.xlabel("dimensions")
plt.ylabel("Time")
```

```python
start = 10
end = 59
print(dumb.iloc[end])
plt.scatter(x = dumb["n"].iloc[start:end], y = dumb["time"].iloc[start:end], color = "red")
plt.scatter(x = bucket["n"].iloc[start:end], y = bucket["time"].iloc[start:end], color = "blue")
plt.scatter(x = quad["n"].iloc[start:end], y = quad["time"].iloc[start:end], color = "green")
plt.scatter(x = kd["n"].iloc[start:end], y = kd["time"].iloc[start:end], color = "black")
plt.legend(["dumb", "bucket", "quad", "kd"],)
plt.xlabel("test points")
plt.ylabel("Time")
```

```python
start = 60
end = 69
print(dumb.iloc[end])
plt.scatter(x = dumb["k"].iloc[start:end], y = dumb["time"].iloc[start:end], color = "red")
plt.scatter(x = bucket["k"].iloc[start:end], y = bucket["time"].iloc[start:end], color = "blue")
plt.scatter(x = quad["k"].iloc[start:end], y = quad["time"].iloc[start:end], color = "green")
plt.scatter(x = kd["k"].iloc[start:end], y = kd["time"].iloc[start:end], color = "black")
plt.legend(["dumb", "bucket", "quad", "kd"],)
plt.xlabel("neighbors")
plt.ylabel("Time")
```
